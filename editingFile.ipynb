```python
import pandas as pd

Book = pd.read_csv("Books.csv")
rating = pd.read_csv("Ratings.csv")
User_rating = pd.read_csv("Users.csv")




print(Book.columns)
print(rating.columns)
print(User_rating.columns)

Book.rename(columns={
    'ISBN': 'book_id',
    'Book-Title': 'title',
    'Book-Author': 'author'
},inplace=True)

rating.rename(columns={
    'User-ID': 'user_id',
    'ISBN': 'book_id',
    'Book-Rating': 'rating'
}, inplace=True)

User_rating.rename(columns={
    'User-ID': 'user_id' 
}, inplace=True)

data = pd.merge(rating, Book, on='book_id')
data = pd.merge(data,User_rating, on='user_id')

data.to_csv("final_output.csv", index=False)
```
