Import Pandas Module and Printing Top 15 Books

```python
import pandas as pd

df = pd.read_csv("final_output.csv")
df.head(15)
```

Taking A Specific Columns And print The df

```python
df = df[['title', 'author', 'rating']]
df
```

Merging the Existing Columns into New Columns

```python
df['tag'] = df['title']+ " " +df['author']
df
```

Cleaning The Data

```python
df = df.dropna()
df = df.drop_duplicates(subset=['title', 'author'])
df = df.reset_index(drop=True)
```

Convert Text into vectors

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=7000, stop_words='english')
vector = vectorizer.fit_transform(df['tag'])
```

Find the index existing position from the title columns

```python
from sklearn.metrics.pairwise import cosine_similarity


idx = df[df['title'] == "The Notebook"].index[0]
```

Find the similarity of number of vectors and creating the list and enumerate the scores to get data in tuple and reverse sorted to get higest similarity to lowest similarity

```python
scores = cosine_similarity(vector[idx], vector).flatten()
scores = list(enumerate(scores))
scores = sorted(scores, reverse=True, key=lambda x:x[1])
```

```python
def recommend(book_name: str):
    # convert it into lower letters
    
    book_name = book_name.lower()
    
    # from the title col convert the small letter and check if the title is ==  to the book name
    matches = df[df['title'].str.lower() == book_name]
     
    # It checks the white space and check the wrong book name example -> harry pttr, not in dataset
    if matches.empty:
        print("Sorry but we can't find your Book please give another book for recomendation")
        return

    
        
    # get the index value
    idx = matches.index[0]

    # get the similarity vector
    scores = cosine_similarity(vector[idx], vector).flatten()

    # convert it list and enumerate the scores
    scores = list(enumerate(scores))

    # sort the scores in descending order
    scores = sorted(scores, reverse=True, key=lambda x:x[1])
    # Give the top 14 books
    scores = scores[1:15]

    # checks the dataset is empty or not if yes then pass this block of code
    if len(scores) == 0:
        print("Sorry we can't Recommend the book")
        return
    
    # give the index value for iteration
    idx_value = []

    # using loop i and append to the blank list
    for i in scores:
        idx_value.append(i[0])

    # Give top books name
    top_books = df.iloc[idx_value]
    # give the data by rating which highly rated by the user 
    final_output = top_books.sort_values(by='rating', ascending=False).head(5)
    
    # return the final_output
    return final_output[['title', 'rating']]
     
    
#  give the book name
book_name = input("Enter your book name you want to read: ")

# call the function
recommend(book_name)
    
```
